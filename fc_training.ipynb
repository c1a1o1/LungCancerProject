```python


import numpy as np
import os
INPUT_FOLDER = '../input/'

#os.environ["CUDA_DEVICE_ORDER"]=PCI_BUS_ID   # see issue #152
os.environ["CUDA_VISIBLE_DEVICES"]="2"

def load(npy):
    return np.load(os.path.join(INPUT_FOLDER,npy))



```

```python

data = load('train_features.npy') # (1397, 224, 2048)
labels = load('y.npy')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten, Convolution2D, MaxPooling2D
```

```python
model = Sequential()
model.add(Flatten(input_shape=(224,2048)))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# 1000, 397 split
model.fit(data, labels, validation_split=397.0/1397.0, 
          nb_epoch=5, batch_size=32)
```
